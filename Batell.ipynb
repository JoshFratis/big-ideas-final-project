To prove which is the best neighborhood,the metric that we are analyzing is safety, as it is important that people feel safe walking on the sidewalks, leaving their car parked on the street, and  other neccesary tasks. To properly analyze the metric of saftey within a neighborhood, knowledge of the area of a neighborhood is necessary. Interpreting the arrests and fires without the area allows for misunderstandings to happen. For example, between two neighborhoods, they may have the same amount of reports, but one is smaller. By reports per area, that one is the less safe neighborhood. This visualization organizes the neighborhoods by largest to smallest areas, in units of acres. 

This visualization is created using a dataset aptly named "Neighborhoods" that was found on the Western Pennsylvania Regional Data Center. This graph informs that Squirrel Hill South is the biggest Pittsburgh neighborhood, while Mt. Oliver is the smallest. If we were to assume equal saftey in all neighborhoods, which is unlikely, Squirrel Hill would have the most incidents and Mt. Oliver would have the least. However, this data cannot completely be interpreted in the context of saftey without the other sub-metrics, only theorized about in hypotheticals.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df= pd.read_csv("neighborhood_area.csv")
area= df.groupby("hood").sum()['acres'].sort_values(ascending=False)

plt.title('Neighborhood Areas by Acres')
plt.xlabel('Neighborhoods')
plt.ylabel('Acres')
area.plot(kind='bar',figsize=(18, 8))
```
