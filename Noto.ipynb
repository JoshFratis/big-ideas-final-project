```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("FireIncidents.csv", sep=",")
df.head()

arrests_by_neighborhood = df['neighborhood'].value_counts()
arrests_by_neighborhood.plot(kind='bar', figsize=(30,15))
plt.title('Number of Fires by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Fires')
plt.show()
```
